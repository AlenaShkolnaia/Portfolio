## Количество обзоров и средняя оценка

```python
books_reviews = ''' WITH books_rewies AS 
                            (SELECT  
                                b.book_id, 
                                b.title, 
                                COUNT(review_id) AS reviews_count
                            FROM books AS b
                            LEFT JOIN reviews AS re ON re.book_id = b.book_id
                            GROUP BY b.book_id),
                            
                        books_ratings AS
                            (SELECT  
                                b.book_id, 
                                ROUND(AVG(ra.rating), 2) AS rating_avg
                            FROM ratings AS ra 
                            LEFT JOIN books AS b ON ra.book_id = b.book_id
                            GROUP BY b.book_id)
                            
                    SELECT * 
                    FROM books_rewies JOIN books_ratings
                    ON books_rewies.book_id = books_ratings.book_id
                    ORDER BY reviews_count DESC
                '''

show_df(books_reviews)
```

В таблице можно увидеть количество отзывов на каждую книгу и ее средний рейтинг. Наибольшее количество отзывов 7 получила книга Сумерки с средней оценкой 3,66.

```python
books_reviews = ''' WITH books_rewies AS 
                            (SELECT  
                                b.book_id, 
                                b.title, 
                                COUNT(review_id) AS reviews_count
                            FROM books AS b
                            LEFT JOIN reviews AS re ON re.book_id = b.book_id
                            GROUP BY b.book_id),
                            
                        books_ratings AS
                            (SELECT  
                                b.book_id, 
                                b.title, 
                                ROUND(AVG(ra.rating), 2) AS rating_avg
                            FROM ratings AS ra 
                            LEFT JOIN books AS b ON ra.book_id = b.book_id
                            GROUP BY b.book_id)
                            
                    SELECT 
                        AVG(reviews_count) AS reviews_avg, 
                        AVG(rating_avg) AS  rating_avg
                    FROM books_rewies JOIN books_ratings
                    ON books_rewies.book_id = books_ratings.book_id
                '''

show_df(books_reviews)
```

В среднем на книгу приходится 2 отзыва. Средняя оценка за книгу 3.9.

## Издательство, которое выпустило наибольшее количество книг

```python
publishers = ''' 
                SELECT 
                    publisher, 
                    COUNT(book_id) AS books_count
                FROM books AS b LEFT JOIN publishers AS p ON b.publisher_id = p.publisher_id
                WHERE num_pages > 50
                GROUP BY publisher
                ORDER BY 2 DESC
                LIMIT 1
                
              '''

show_df(publishers)
```

Наибольшее количество книг с объемом страниц более 50 выпустило издательство под названием Penguin Books.

## Автор с самой высокой средней оценкой 

```python
authors =       '''       
                    WITH rating_top AS
                        (SELECT 
                               b.book_id, 
                               b.author_id,
                               b.title, 
                               COUNT(b.book_id) AS rating_count, 
                               AVG(ra.rating) AS rating_avg
                               
                     FROM books AS b JOIN ratings AS ra ON ra.book_id = b.book_id
                     GROUP BY b.book_id
                     HAVING COUNT(b.book_id) > 50
                     ORDER BY 4 DESC),
                        
                    rating_avg AS
                        (SELECT 
                        author, 
                        title, 
                        ROUND(rating_avg, 2) AS rating_avg
                    FROM rating_top JOIN authors AS au ON au.author_id = rating_top.author_id
                    ORDER BY 3 DESC)
                    
                    SELECT
                        rating_avg.author,
                        AVG(rating_avg)
                    FROM  rating_avg
                    GROUP BY author
                    LIMIT 1
                      

                '''

show_df(authors)
```

Автор с более высокой средней оценкой из популярных книг (тех, у которых количество оценок превышает 50 за книгу) - Джоан Роулинг.

## Среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
users =       '''      
                        WITH 
                        users AS
                            (SELECT username
                            FROM  ratings AS ra 
                            GROUP BY username
                            HAVING COUNT(rating_id) > 50), 
                            
                        reviews AS
                            (SELECT re.username, COUNT(re.review_id) AS reviews_count
                            FROM reviews AS re
                            JOIN users ON re.username = users.username
                            GROUP BY re.username)
                        
                        SELECT ROUND(AVG(reviews_count)) AS reviews_avg
                        FROM reviews
                        
                '''
show_df(users)
```

В среднем активные пользователи (которые ставят более 50 оценок) пишут 24 обзора на книги.
